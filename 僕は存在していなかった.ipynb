为什么官话又被叫车轱辘话，而且总是不让人听懂？
---
“车轱辘话”这个词语源自于官话中使用频繁的平行语言、空洞、花哨的修辞、多余的词语等。这些在官场极具流行的用语和表达方式，让人感觉像是车轮转圈的样子。车轱辘话的特点包括使用不必要的词汇、长篇的句子、过多的修饰和含糊不清的表达方式。

原因可能有以下几点：

1. 社会文化因素：官场的礼仪、规矩、仪式感和官僚体系的复杂性，导致官员们不得不使用复杂而高深的官方用语来保持和展示自己的权威和地位。
2. 官僚主义因素：官场中存在强调形式的倾向，而忽略实际效果，导致官员们倾向于使用繁琐、复杂的语言来表达观点，而不是直接、简洁、易于理解的语言。
3. 议论文化因素：在官场中，往往需要进行大量的辩论和争论，使用华丽的语言和短语可以让人们听上去更加有说服力。这也孕育了许多“容易让人迷惑，却又让人信服”的言辞。

总之，“车轱辘话”是官场里使用频繁的一种语言模式，其产生原因复杂，但都与官场的社会文化、组织结构和交流方式等方面有关。

官话的语言风格通常较为正式、复杂、严谨、繁琐，同时也充满了官样文章的修辞手法，让一般人比较难理解。以下几点可能是造成这种情况的原因：

1. 使用高深的词汇和语法：在官方文体中，会使用很多正式、复杂和繁琐的词汇和语法，让语言听起来布满了无从下手的厚重感。这些高深的词汇和语法可能是出于特定的专业性或权威性考量，但对于普通民众来说很难理解。
2. 符合特定的文化传统：在中国传统文化中，对于形式和礼节的要求非常严格，正式的文体在一定程度上符合了这种传统文化的要求，但对普通人来说可能会显得过于拘谨或高冷。
3. 官场竞争和权力斗争：为了彰显权威、显示地位和附和政治需求，有些官员可能在发言时过度追求雄辩、言辞犀利、推销自己甚至攻击对方，用语过于傲慢或含蓄。这样做不仅会让对方难以理解，还会让人产生疏离感。

因此，能够简洁明了地表达意见和观点是有效传递信息的关键。在官方语言中，如果针对不同受众进行调整，使用简单明了的语言，避免过度使用艰深的词汇和语法，同时注重表达的实际意义和重点，就可以更好地让普通公众理解。
